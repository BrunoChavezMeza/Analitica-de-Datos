```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_excel("Datos de Facturación_Limpio.xlsx")

gc= pd.read_excel("Gastos y costos_2020_Limpio.xlsx")
gc1= pd.read_excel('Gastos y costos_2021_Limpio.xlsx')
gc2= pd.read_excel('Gastos y costos_2022_Limpio.xlsx')
gc3= pd.read_excel('Gastos y costos_2023_Limpio.xlsx')
```

```python

import pandas as pd
gc_cuantitativas = gc.iloc[:,[0,9,10,11,12,13,14,15,19]]
gc1_cuantitativas = gc1.iloc[:,[0,9,10,11,12,13,14,15]]
gc2_cuantitativas = gc2.iloc[:,[0,9,10,11,12,13,14,15]]
gc3_cuantitativas = gc3.iloc[:,[0,9,10,11,12,13,14,15]]


print(gc_cuantitativas)
###################

gc_cualitativas = gc.iloc[:,[1,2,3,4,5,6,7,8,16,17,18]]
gc1_cualitativas = gc.iloc[:,[1,2,3,4,5,6,7,8,16,17,18]]
gc2_cualitativas = gc.iloc[:,[1,2,3,4,5,6,7,8,16,17,18]]
gc3_cualitativas = gc.iloc[:,[1,2,3,4,5,6,7,8,16,17,18]]


```

```python
gc_cualitativas.head()
```

```python

y=gc_cuantitativas
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() -3*y.std()

y1=gc1_cuantitativas
Limite_Superior1= y.mean() + 3*y.std()
Limite_Inferior1= y.mean() -3*y.std()

y2=gc2_cuantitativas
Limite_Superior2= y.mean() + 3*y.std()
Limite_Inferior2= y.mean() -3*y.std()

y3=gc3_cuantitativas
Limite_Superior3= y.mean() + 3*y.std()
Limite_Inferior3= y.mean() -3*y.std()
```

```python
#Obtenemos datos y los outliers se conveirten en nulos en el DataFrame

gc_cuantitativas_out= gc_cuantitativas[(y<=Limite_Superior)&(y>=Limite_Inferior)]
gc_cuantitativas_out

gc1_cuantitativas_out= gc1_cuantitativas[(y<=Limite_Superior1)&(y>=Limite_Inferior1)]
gc1_cuantitativas_out

gc2_cuantitativas_out= gc2_cuantitativas[(y<=Limite_Superior2)&(y>=Limite_Inferior2)]
gc2_cuantitativas_out

gc3_cuantitativas_out= gc3_cuantitativas[(y<=Limite_Superior3)&(y>=Limite_Inferior3)]
gc3_cuantitativas_out

```

```python
#Reemplazamos valores atípicos (nulos) del data frame con "mean"
#Realizamos una copia del dataframe
data_cuantitativas_out_clean=gc_cuantitativas_out.copy()
data_cuantitativas_out_clean=data_cuantitativas_out_clean.fillna(round(gc_cuantitativas_out.mean(),1))

data_cuantitativas_out_clean1=gc1_cuantitativas_out.copy()
data_cuantitativas_out_clean1=data_cuantitativas_out_clean1.fillna(round(gc1_cuantitativas_out.mean(),1))

data_cuantitativas_out_clean2=gc2_cuantitativas_out.copy()
data_cuantitativas_out_clean2=data_cuantitativas_out_clean2.fillna(round(gc2_cuantitativas_out.mean(),1))

data_cuantitativas_out_clean3=gc3_cuantitativas_out.copy()
data_cuantitativas_out_clean3=data_cuantitativas_out_clean3.fillna(round(gc3_cuantitativas_out.mean(),1))
```

```python
GastosyCostos2020= pd.concat([data_cuantitativas_out_clean,gc_cualitativas], axis=1)
GastosyCostos2020

GastosyCostos2021= pd.concat([data_cuantitativas_out_clean1,gc1_cualitativas], axis=1)
GastosyCostos2021

GastosyCostos2022= pd.concat([data_cuantitativas_out_clean2,gc2_cualitativas], axis=1)
GastosyCostos2022

GastosyCostos2023= pd.concat([data_cuantitativas_out_clean3,gc3_cualitativas], axis=1)
GastosyCostos2023
```

```python
# prompt: exportar un dataframe a excel

GastosyCostos2020.to_excel("GastosyCostos2020_out.xlsx", sheet_name="Gastos y Costos 2020_out")
GastosyCostos2021.to_excel("GastosyCostos2021_out.xlsx", sheet_name="Gastos y Costos 2021_out")
GastosyCostos2022.to_excel("GastosyCostos2022_out.xlsx", sheet_name="Gastos y Costos 2022_out")
GastosyCostos2023.to_excel("GastosyCostos2023_out.xlsx", sheet_name="Gastos y Costos 2023_out")

```

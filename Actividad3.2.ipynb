```python
gc_bueno= pd.read_excel('Gastos y costos 20-23.xlsx',sheet_name='2020',skiprows=5)
gc_bueno
```

```python
gc_bueno.info()
```

```python
gc_bueno.head()

gc_bueno['TC']=gc_bueno['TC'].fillna(round(gc_bueno['TC'].mean(),1))
gc_bueno['IMPORTE']=gc_bueno['IMPORTE'].fillna(round(gc_bueno['IMPORTE'].mean(),1))
gc_bueno['IVA']=gc_bueno['IVA'].fillna(round(gc_bueno['IVA'].mean(),1))
gc_bueno['RET ISR']=gc_bueno['RET ISR'].fillna(round(gc_bueno['RET ISR'].mean(),1))
gc_bueno['RET IVA']=gc_bueno['RET IVA'].fillna(round(gc_bueno['RET IVA'].mean(),1))
gc_bueno['TOTAL MX']=gc_bueno['TOTAL MX'].fillna(round(gc_bueno['TOTAL MX'].mean(),1))
gc_bueno['TOTAL SAT']=gc_bueno['TOTAL SAT'].fillna(round(gc_bueno['TOTAL SAT'].mean(),1))
gc_bueno['Columna1']=gc_bueno['Columna1'].fillna(round(gc_bueno['Columna1'].mean(),1))

gc_bueno['GASTO'] = gc_bueno['GASTO'].fillna(method="bfill")
gc_bueno['POLIZA'] = gc_bueno['POLIZA'].fillna(method="bfill")

columns_to_fill = ['FOLIO', 'UUID', 'RFC', 'PROVEEDOR', 'TIPO GASTO', 'DESCRIPCION', 'TIPO', 'STATUS','POLIZA']
gc_bueno[columns_to_fill] = gc_bueno[columns_to_fill].fillna(method="ffill")


gc_bueno.isnull().sum()
```
